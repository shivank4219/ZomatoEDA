checking for null values in any column

```python
data.isna().sum()
```

```python
data['rate']=data['rate'].replace('NEW',np.nan)
data['rate']=data['rate'].replace('-',np.nan)
data.dropna(how='any',inplace=True)
```

```python
data.isna().sum()
```

```python
data['rate']
```

```python
data['rate']=data['rate'].replace('[ ]','',regex=True)
data['rate']=data['rate'].replace('/5','',regex=True)
data['rate']=data['rate'].apply(lambda r:float(r))
```

```python
data['rate'].head()
```

```python
data['approx_cost(for two people)']=data['approx_cost(for two people)'].replace(',','',regex=True)
```

```python
data['approx_cost(for two people)']
```

```python
slices=[((data.rate>=1.5) & (data.rate<2)).sum(),
        ((data.rate>=2) & (data.rate<2.5)).sum(),
        ((data.rate>=2.5) & (data.rate<3)).sum(),
        ((data.rate>=3.0) & (data.rate<3.5)).sum(),
        ((data.rate>=3.5) & (data.rate<4)).sum(),
        ((data.rate>=4) & (data.rate<4.5)).sum(),
        ((data.rate>=4.5) & (data.rate<5)).sum()
       ]
labels=['1.5-2','2-2.5','2.5-3','3-3.5','3.5-4','4-4.5','4.5-5']
colors = ['Red','blue','Green','black','orange','pink',"cyan"]
explode = [0.5,0,0,0.5,0,0,0]
plt.pie(slices,colors=colors,labels=labels,autopct='%1.02f%%',explode = explode)
fig = plt.gcf()
plt.title("Percentage of Restaurants according to their ratings")

fig.set_size_inches(8,8)
plt.show()
```

Analysis of restaurant based on location

```python
#total no of unique location
data['location'].nunique()
```

```python
print("top 10 loction with max no. of restaurants")
dt=data['location'].value_counts().head(10)
dt
```

```python
plt.figure(figsize=(20,8))
sns.barplot(x=dt.index,y=dt,palette='dark')
plt.title("No. of restaurant at famous places ")
```

```python
print("location with max no. of restaurant")
for x,y in dt.items():
    if y==max(dt):
        print(x)
```

```python
print("location with min no. of restaurants")
dt=data['location'].value_counts().sort_values(ascending=True)
for x,y in dt.items():
    if y==min(dt):
        print(x)
```

```python
plt.figure(figsize=(20,40))
dt.plot(kind='barh',fontsize=20)
plt.xlabel('restaurant count',fontsize=30)
plt.ylabel('name of places',fontsize=30,color='black')
plt.title('location vs restaurant count',fontsize=40)
```
